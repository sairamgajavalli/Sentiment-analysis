```python
import pandas as pd
import re
```

```python
import nltk
```

```python
hmid=pd.read_csv('hm_train.csv')
```

```python
hmid_100=hmid[0:10000]
```

```python
pd.set_option("display.max_columns",100)
pd.set_option("display.max_row",100)
pd.set_option("max_colwidth",150)

hmid_100
```

```python
import nltk
from nltk.corpus import stopwords
stop_1=stopwords.words('english')
stop_1
```

```python
hmid_100['predicted_category']=hmid_100['predicted_category'].replace({'affection':1,'leisure':0,'enjoy_the_moment':2,'exercise':3,'bonding':4,'nature':5,'achievement':6})
hmid_100
```

```python
hmid_100.isnull().sum()
```

```python
hmid_100['cleaned_hm'].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
vectorizer
```

```python
x=vectorizer=CountVectorizer(max_df=.85,stop_words=stop_1,max_features=60000,token_pattern='[a-zA-Z][a-zA-Z]{2,}')
x=vectorizer.fit_transform(hmid_100['cleaned_hm'])
x.toarray()
```

```python
feature_names=vectorizer.get_feature_names()
feature_names
```

```python
feature_n=pd.DataFrame(x.toarray(),columns=vectorizer.get_feature_names())
feature_n
```

```python
x=feature_n
x.shape
```

```python
y=hmid_100['predicted_category']
y.dtypes
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=25,random_state=1)
```

```python
x_train.shape
```

```python
x.shape
```

```python
from sklearn import tree
ctree=tree.DecisionTreeClassifier()
ctree
```

```python
from sklearn import metrics
ctree.fit(x_train,y_train)
y_train_pred=ctree.predict(x_train)
metrics.accuracy_score(y_train,y_train_pred)
```

```python
y_test_pred=ctree.predict(x_test)
metrics.accuracy_score(y_test,y_test_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { "criterion"      : ['gini', 'entropy'],    
               "max_features"   : [2, 3, 4, 5, 6, 7 ],             
               "splitter"       : ['best', 'random'],  
               "min_samples_split"  : [3, 5, 10 ],
               "max_depth"          : [2,3,4,5] ,
               "class_weight"      : [{0:1, 1:2}, {0:1, 1:5}]
               }

```

```python
optimized_tree=GridSearchCV(ctree,param_grid,scoring='f1',verbose=2,cv=5)
```

```python
optimized_tree.fit(x_train,y_train)
```

```python
optimized_tree.best_params_
```

```python
optimized_tree.best_score_
```

```python
final_model = optimized_tree.best_estimator_
final_model
```

```python
y_test_pred = final_model.predict(x_test)
metrics.accuracy_score(y_test,y_test_pred)
```

```python
metrics.recall_score(y_test,y_test_pred,pos_label='positive',average='macro')
```

```python
metrics.precision_score(y_test,y_test_pred,pos_label='positive'
                                           ,average='macro')
```
